```python
import pandas as pd
import os
import sys
import numpy as np
SKIP_FILES = ['cmds']
NEWLINE='\n'
HAM = 'ham'
SPAM = 'spam'

SOURCES = [
    ('data/spam',        SPAM),
    ('data/easy_ham',    HAM),
    ('data/hard_ham',    HAM),
    ('data/beck-s',      HAM),
    ('data/farmer-d',    HAM),
    ('data/kaminski-v',  HAM),
    ('data/kitchen-l',   HAM),
    ('data/lokay-m',     HAM),
    ('data/williams-w3', HAM),
    ('data/BG',          SPAM),
    ('data/GP',          SPAM),
    ('data/SH',          SPAM)
]

%matplotlib inline
```

```python

def read_files(path):

    for root, dir_name, filename in os.walk(path):
        
        for f in filename:
            content="\n"
            if filename not in SKIP_FILES:

                filen=os.path.join(root,f)
                if os.path.isfile(filen):
                    with open(filen,encoding='latin-1') as d:
                        content=d.read()
                yield filen, content
    
```

```python
def build_dataframe(path,classi):
    row=[]
    index=[]

    for i, (path, con) in enumerate(read_files(path)):
        #print(i)
        #print(path)
        #print(con)
        row.append({'text': con, 'class': classi})
        index.append(path)
    dataf=pd.DataFrame(row, index=index)
    return dataf
    
```

```python
df=build_dataframe("data/spam","spam")
```

```python
def load_data():
    total=len(SOURCES)+1
    i=1;
    data = pd.DataFrame({'text': [], 'class': []})
    for path, classification in SOURCES:
        dr=build_dataframe(path,classification)
        data = data.append(dr)
        print(total,"/",i,"done")
        i+=1
    #data = data.reindex(numpy.random.permutation(data.index))
    return data
```

```python
data=load_data()
```
